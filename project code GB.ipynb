```python
#importing necessary packages
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import polars as pl
```

```python
nodes = pl.read_csv(
    "nodes.csv",
    separator=",",
    encoding="utf8-lossy",  # tolerate encoding errors and BOMs
    ignore_errors=True
) #this uploads the data as a dataframe called "nodes" and enables us to filter the data and drop null (NA) values
nodes = nodes.with_columns([
    pl.col(pl.Int64).cast(pl.Utf8)
]) #changes the data from int64 to str so that the filtering code below can work

nodes.drop_nulls() #drops nulls
filtered_values = ["1", "2", "3", "4"]

nodes = nodes.filter(~pl.col("race").is_in(filtered_values)) #this would filter out the above values in the race column, so that the only remaining nodes are those of race 5.
#the "race" text can be replaced with the other column names, and the filtered values can be changed as needed for the project
print(nodes)
```
